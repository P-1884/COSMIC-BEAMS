```python
# db = pd.read_csv('./databases/real_paltas_population_TP_10390_FP_48610_Spec_990_P_0.2.csv')
db = pd.read_csv('./databases/real_paltas_population_TP_100000_FP_100000_Spec_10000_P_0.5.csv')
hist_dict = {'bins':np.linspace(-4,4,41),'density':True,'edgecolor':'k','alpha':0.5}
pl.hist((db['zL_obs']-db['zL_true'])/db['sigma_zL_obs'],**hist_dict,label='zL',fill='darkblue')
pl.hist(((db['zS_obs']-db['zS_true'])/db['sigma_zS_obs']),**hist_dict,label='zS',fill='darkred')
X_plot = np.linspace(-3,3,100)
pl.plot(X_plot,norm.pdf(X_plot))
pl.legend()
pl.xlabel('(z_obs - z_true)/sigma_z_obs')
pl.ylabel('Probability Density')
pl.show()
print('zL',kstest((db['zL_obs']-db['zL_true'])/db['sigma_zL_obs'],'norm').pvalue)
print('zS',kstest((db['zS_obs']-db['zS_true'])/db['sigma_zS_obs'],'norm').pvalue)
```
